```python
#Connecting to MongoDB Atlas server.  Accessing the ML_Speed_Dating database

client = pymongo.MongoClient(Mongouri)
db = client.ML_Speed_Dating
```

```python
#function to convert 3 columns from string to int
#mn_sat, tution, income

def string_to_number(val):
    return int(''.join(filter(str.isdigit, val)))
```

```python
#reading in csv used to run ML and prepping to load into MongoDB
#convert to df, dropped NA
#mapped through 3 columns to ensure no string values

df = pd.read_csv("./Data/3_ML_Dry_Run.csv")
df = df.dropna()
df["mn_sat"] = df["mn_sat"].map(string_to_number)
df["tuition"] = df["tuition"].map(string_to_number)
df["income"] = df["income"].map(string_to_number)


df_list = df.to_dict("record")
df.head()
```

```python
#loading general data collection to MongoDB


for row in df_list:
    db.General_Data.insert_one(row)
    
```

```python
#Prepping features of importance to load to MongoDB
#csv to df
#df to list of dict

df2 = pd.read_csv("./Data/Combined_features_Importance.csv")
df2_list = df2.to_dict("record")
df2.head()
```

```python
#loading combined features of importance to MongoDB

for row in df2_list:
    db.Combined_features_Importance.insert_one(row)
    
```
